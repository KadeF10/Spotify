```python
import requests 
import pandas as pd
url= 'https://raw.githubusercontent.com/KadeF10/Spotify/main/data.csv'
df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
print(df.dtypes)
```

```python
df.corr()
```

```python
sns.regplot(x="energy", y="tempo", data=df)
plt.ylim(0,)
```

```python
df.describe()
```

```python
df['acousticness'].plot(kind='hist', figsize=(8, 5))
plt.title('Accousticness')
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
df['danceability'].plot(kind='hist', figsize=(8, 5))
plt.title('Danceability')
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
df['energy'].plot(kind='hist', figsize=(8, 5))
plt.title('energy')
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
